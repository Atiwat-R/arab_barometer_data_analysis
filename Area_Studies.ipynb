```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Import Arab Barometer data
df = pd.read_csv('./Arab_Barometer_Wave_6_Part_3_ENG_RELEASE.csv')

# Display DataFrame
display(df)
```

```python
# Replace placeholder numbers with country names

country_name = {
    1 : "Algeria",
    7 : "Iraq",
    8 : "Jordan",
    10 : "Lebanon",
    11 : "Libya",
    13 : "Morocco",
    21 : "Tunisia"
}

df['COUNTRY'] = df['COUNTRY'].map(country_name)
```

```python
# This is the new column

df[['COUNTRY']]
```

```python
# Fix column names

new_col = {
  "COUNTRY" : "country",
  "Q1001" : "age",
  "Q1002" : "gender",
  "Q1003" : "education",
  "Q1016" : "income",
  "Q1012" : "religion",
  "Q1012B" : "ethnicity",
  "Q609" : "religiousity",
  "Q602_1" : "tolerance_religion",
  "Q602_2" : "tolerance_race",
  "Q602_3" : "tolerance_immig",
  "Q201A_1" : "trust_gov",
  "Q201A_2" : "trust_court",
  "Q204A_3" : "satisfy_gov",
  "Q204A_1" : "satisfy_edu",
  "Q204A_2" : "satisfy_healthsys",
  "Q521_1" : "free_express",
  "Q521_2" : "free_media",
  "Q521_4" : "free_demo",
  "Q204_11" : "security",
  "Q204_20" : "price",
  "Q204_25" : "covid19",
  "Q210" : "corruption",
  "Q211" : "anti_corruption"
}

df.rename(columns=new_col, inplace=True)
```

```python
# Keep only the columns we want, now in df2

df2 = df[[
    "country",
    "age",
    "gender",
    "education",
    "income",
    "religion",
    "ethnicity",
    "religiousity",
    "tolerance_religion",
    "tolerance_race",
    "tolerance_immig",
    "trust_gov",
    "trust_court",
    "satisfy_gov",
    "satisfy_edu",
    "satisfy_healthsys",
    "free_express",
    "free_media",
    "free_demo",
    "security",
    "price",
    "covid19",
    "corruption",
    "anti_corruption"
]]

df2
```

```python
# Libraries for plotting
# A lot of columns came with very limited varieties (ex. 1-4) so graphs for categorical data such as barplot are more suited 

import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
# Since there's missing values (NaN) let's visualize how much it is
# Credit: https://datavizpyr.com/visualizing-missing-data-with-seaborn-heatmap-and-displot/

plt.figure(figsize=(10,6))
sns.heatmap(df2.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})

# plt.show("nan_heatmap.png", dpi=100) # Save png
```

```python
# From the look of it, most columns have available data. 
# Only ethnicity and anti-corruption is significantly missing.
```

```python
# I wonder if there's some correlation between missing ethnicity data and countries...

eth_df = df[["country", "ethnicity"]]

# Looks like only 3 countries out of 7 are missing ethnicity data.
sns.barplot(data=eth_df, x="country", y="ethnicity")
```

```python
# I presume that there is correlation between freedom in media & expression?
free_df = df2.copy(deep=True)
free_df = free_df.drop(free_df[free_df["free_express"] > 97].index) # drop col with unknown response
free_df = free_df.drop(free_df[free_df["free_media"] > 97].index) # drop col with unknown response

free_df['free_express'] = free_df['free_express'].map({1:4, 2:3, 3:2, 4:1}) # Turn values around for easier graphs
free_df['free_media'] = free_df['free_media'].map({1:4, 2:3, 3:2, 4:1}) # Turn values around for easier graphs

# Indeed there is linear correlation.
sns.lineplot(data=free_df, x="free_media", y="free_express")

# People believe freedom in expression and media goes hand-to-hand.
```

```python
# Let's see more details in regards to each country.

sns.lineplot(data=free_df, x="free_media", y="free_express", hue="country", err_style="bars")
```

```python
# How about satisfaction in health & education?

# I presume that there is correlation between freedom in media & expression?
eduhealth_ef = df2.copy(deep=True)
eduhealth_ef = eduhealth_ef.drop(eduhealth_ef[eduhealth_ef["satisfy_edu"] > 97].index) # drop col with unknown response
eduhealth_ef = eduhealth_ef.drop(eduhealth_ef[eduhealth_ef["satisfy_healthsys"] > 97].index) # drop col with unknown response

eduhealth_ef["satisfy_edu"] = eduhealth_ef["satisfy_edu"].map({1:4, 2:3, 3:2, 4:1}) # Turn values around for easier graphs
eduhealth_ef["satisfy_healthsys"] = eduhealth_ef["satisfy_healthsys"].map({1:4, 2:3, 3:2, 4:1}) # Turn values around for easier graphs

# For some countries like Libya, satisfaction in education is high, but satisfaction in healthcare is mediocre.
sns.lineplot(data=eduhealth_ef, x="satisfy_edu", y="satisfy_healthsys", hue="country", err_style="bars")
```

```python
# Strange case for Lebanon.
# Those who are highly satisfied with healthcare are hardly satisfied with education,
# yet those less satisfied with heathcare are a little more satisfied with education.
# These are just on average, of course.

sns.lineplot(data=eduhealth_ef, x="satisfy_healthsys", y="satisfy_edu", hue="country", err_style="bars")
```

```python
# Next, how about age and religiousity for each religion?

agereligty = df2.copy(deep=True)
agereligty = agereligty.drop(agereligty[agereligty["age"] > 100].index) # drop col with unknown response
agereligty = agereligty.drop(agereligty[agereligty["religiousity"] > 97].index) # drop col with unknown response
agereligty = agereligty.drop(agereligty[agereligty["religion"] > 10].index) # drop col with unknown response

agereligty["religiousity"] = agereligty["religiousity"].map({1:3, 3:1}) # Turn values around for easier graphs
agereligty["religion"] = agereligty["religion"].map({1:"Muslim", 2:"Christian", 3:"Unspecified", 4:"Atheist", 5:"Others"}) # Turn values around for easier graphs

# Seems very sporatic...
sns.lineplot(data=agereligty, x="age", y="religiousity", hue="religion")
```
